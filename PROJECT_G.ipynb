Finding the most frequent word patterns

```python
import pandas as pd
from collections import Counter
import re

# df = pd.DataFrame(data)

def extract_word_patterns(row, pattern_length=1):
    text = row['Ticket Description']
    words = re.findall(r'\b\w+\b', text.lower())  # Tokenize the text into words
    patterns = [tuple(words[i:i+pattern_length]) for i in range(len(words) - pattern_length + 1)]
    return patterns

pattern_length = 1

df['Word Patterns'] = df.apply(lambda row: extract_word_patterns(row, pattern_length), axis=1)

all_patterns = [item for sublist in df['Word Patterns'].tolist() for item in sublist]
pattern_counts = Counter(all_patterns)

most_frequent_patterns = pattern_counts.most_common(5)  # Change 5 to the desired number of top patterns

print("Most frequent word patterns:")
print(most_frequent_patterns)
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import word_tokenize
```

```python
# Preprocessing function with Bigrams
def preprocess_text(text, stemming=True):
    # Lowercasing
    text = text.lower()

    # Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)

    # Tokenization
    tokens = word_tokenize(text)

    # Stopword removal
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word not in stop_words]

    # Remove unnecessary words (customize this list based on your needs)
    unnecessary_words = set(['please', 'assist', 'appreciate', 'issue', 'tried', 'step', 'mention', 'persist'])
    filtered_tokens = [word for word in filtered_tokens if word not in unnecessary_words]

    # Stemming or Lemmatization
    if stemming:
        stemmer = PorterStemmer()
        preprocessed_tokens = [stemmer.stem(word) for word in filtered_tokens]
    else:
        lemmatizer = WordNetLemmatizer()
        preprocessed_tokens = [lemmatizer.lemmatize(word) for word in filtered_tokens]

    # Join the tokens back to form preprocessed text
    preprocessed_text = " ".join(preprocessed_tokens)

    return preprocessed_text

# Apply preprocessing with Bigrams to each text column
text_columns = ['Ticket Description']  # Add more column names if needed
for col in text_columns:
    data_encoded[col] = data_encoded[col].apply(preprocess_text)
```

Feature Engineering

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Assuming you have the data in a DataFrame named 'data_encoded'
# Perform TF-IDF vectorization on 'Ticket Description'
tfidf_vectorizer = TfidfVectorizer()
X_tfidf = tfidf_vectorizer.fit_transform(data_encoded['Ticket Description'])

# Combine the TF-IDF representation with the one-hot encoded 'Ticket_Subject' columns
X_combined = pd.concat([data_encoded.drop('Ticket Description', axis=1), pd.DataFrame(X_tfidf.toarray())], axis=1)
```

```python
X_combined.info()
```

Model Selection and Training

```python
# Convert all column names in X_combined to strings
X_combined.columns = X_combined.columns.astype(str)

# Define the target variable 'y' (the one-hot encoded 'Ticket_Type' columns)
y = data_encoded[['Ticket_Type_Billing inquiry', 'Ticket_Type_Cancellation request', 'Ticket_Type_Product inquiry', 'Ticket_Type_Refund request', 'Ticket_Type_Technical issue']]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_combined, y, test_size=0.2, random_state=60)

# Initialize the RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=500 ,random_state=60)

# Train the classifier on the training data
classifier.fit(X_train, y_train)

# Make predictions on the testing data
y_pred = classifier.predict(X_test)

```

Model Evaluation

```python
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)*100
print("Model Accuracy:", accuracy)

# Get the classification report for more detailed evaluation
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```
